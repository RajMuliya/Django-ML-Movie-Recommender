```python
import pandas as pd
import json
from bs4 import BeautifulSoup
import requests
import re
from imdb import Cinemagoer
import time
import numpy as np
import ast
```

```python
df=pd.read_csv("MoMovie.csv")
```

```python
df.sample(5)
```

```python
df=df.drop(['web-scraper-order', 'web-scraper-start-url', 'link'], axis=1)
```

```python
def remove_number(text):
    """
    Removes leading numbers followed by a dot and space from the text.
    Example: "1. Example Title" -> "Example Title"
    """
    return re.sub(r'^\d+\.\s*', '', text)
```

```python
df['Movie_Name'] = df['Movie_Name'].apply(remove_number)
```

```python
df['Duration'] = df['Duration'].replace(np.nan , "Unknown")
df['Rating'] = df['Rating'].replace(np.nan , "Unknown")
df['Overview'] = df['Overview'].replace(np.nan , "Unknown")
```

```python
df.dropna(inplace=True)
```

```python
df.sample(5)
```

```python
def extract_ids(link_string):
    # Use regex to find all occurrences of 'tt' followed by digits
    ids = re.findall(r'tt\d+', link_string)
    return ', '.join(ids)
```

```python
df['imdb_id'] = df['link-href'].apply(extract_ids)
```

```python
def extract_ids(link_string):
    # Use regex to find all occurrences of 'tt' followed by digits, then remove 'tt'
    ids = re.findall(r'tt(\d+)', link_string)  # Capture only the digits after 'tt'
    # Join the list of numeric IDs into a single string, separated by commas
    return ', '.join(ids)
```

```python
df['movie_id'] = df['imdb_id'].apply(extract_ids)
```

```python
df=df.drop(['link-href'], axis=1)
df
```

```python
ia = Cinemagoer()
```

```python
directors_list = []
cast_list = []
genres_list = []
```

```python
for movie_id in df['movie_id']:
    movie = ia.get_movie(movie_id)
    director_names = movie.get('directors', [])  # Use .get() to avoid KeyError

    if director_names:  # If the list is not empty
        # Join director names into a single string
        director = ', '.join(director['name'] for director in director_names)
    else:  # If no directors are found
        director = "unknown"  # Assign "unknown"

    directors_list.append(director)  # Append the string to the list

    cast_members = movie.get('cast', [])
    if cast_members:  # If the list is not empty
        # Limit the number of cast members if desired (e.g., top 5)
        cast = ', '.join(actor['name'] for actor in cast_members[:5])
    else:  # If no cast is found
        cast = "unknown"  # Assign "unknown"

    cast_list.append(cast)  # Append the string to the cast list

    # Fetch genres
    genres = movie.get('genres', [])
    if genres:  # If the list is not empty
        genre = ', '.join(genres)
    else:  # If no genres are found
        genre = "unknown"  # Assign "unknown"

    genres_list.append(genre)  # Append the string to the genres list
    time.sleep(0.3)  # Sleep to avoid overwhelming the API
```

```python
gen_cast_dir = pd.DataFrame({
    'Directors': directors_list,
    'Casts': cast_list,
    'Genres': genres_list
})

gen_cast_dir
```

```python
updated_df = df.assign(**gen_cast_dir)
```

```python
updated_df.sample(5)
```

```python
updated_df.to_csv('Processed_2024_updated.csv')
```

```python
def remove_alpha_or_nan(year):
    # Use regular expression to remove all alphabetic characters
    cleaned_year = re.sub(r'[A-Za-z]', '', year)
    
    # If the result is an empty string, return NaN
    if cleaned_year == '':
        return 0
    return cleaned_year
```
